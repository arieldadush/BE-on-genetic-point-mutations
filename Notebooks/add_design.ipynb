```python
#define function to swap columns
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df

#swap points and rebounds columns
sift_df = swap_columns(sift_df, 'RNA off-target hits -minIdentity=85', '_variantId')
sift_df
```

add CodonImprov          

```python
final_df=final_df.rename({'new_codon_seq':'Reference_Triade','new_alt_three_nuc':'Mutant_Triade','List_of_editing':'Edited_Triade'}, axis=1)
final_df
```

```python
from Bio.Seq import Seq
from Bio.SeqUtils import seq3
def get_amino_acid(codon):
    if isinstance(codon, str):
        tran_cod = Seq(codon).translate()
        seq_three =  seq3(tran_cod)
        return seq_three
    else:
        return np.nan

final_df['ref_AA'] =final_df.apply(lambda x: get_amino_acid(x.Reference_Triade), axis=1)
final_df['mut_AA'] =final_df.apply(lambda x: get_amino_acid(x.Mutant_Triade), axis=1)
final_df['edit_AA'] =final_df.apply(lambda x: get_amino_acid(x.Edited_Triade), axis=1)
#final_df.loc[~final_df['Reference Amino Acid'].isna(),'Reference Amino Acid'] = seq3(final_df['Reference Amino Acid'])
final_df
```

```python
final_df['CodonImprov'] = 'not relevant'
final_df.loc[final_df['EditingOption'].str.contains('AA improvement'),'CodonImprov'] = '(' +final_df['Reference_Triade'] + ')' +final_df['Mutant_Triade'] + '>' + final_df['Edited_Triade'] + ' ; ' +'(' +  final_df['ref_AA']+')' + final_df['mut_AA']+'>' +final_df['edit_AA'] 
final_df
```
