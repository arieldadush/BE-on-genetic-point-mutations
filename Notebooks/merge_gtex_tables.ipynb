```python
# import libraries
import pandas as pd
import numpy as np
```

I downloaded the TPM file from the GTEx site :
https://storage.googleapis.com/gtex_analysis_v8/rna_seq_data/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm.gct.gz

```python
#read the GTEx tpm file as tsv 
tpm_path = "/private6/Projects/dadush/SIFT_try2/erez_request/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm.gct"
# skip first 2 in order to avoid problems with the amount of fields in the table 
tpm_df = pd.read_csv(tpm_path, sep='\t',skiprows=2)
tpm_df
```

```python
# check for duplicates  - there isn't duplications in Name col, but in Description col there is 1608 
tpm_df[tpm_df['Description'].duplicated()]
#tpm_df[tpm_df['Description']=='Y_RNA']
```

```python
#To avoid duplicates just because of the gene name, I connected the gene name and its ID in the ensemble
tpm_df['Description'] = tpm_df['Description'] +'#' + tpm_df['Name']
tpm_df
```

```python
# set Description column as index for transpose later
tpm_reset_index_tpm_df = tpm_df.set_index('Description')
tpm_reset_index_tpm_df
```

```python
# drop 'Name' column
tpm_reset_index_tpm_without_name_df = tpm_reset_index_tpm_df.drop('Name', axis=1)
tpm_reset_index_tpm_without_name_df
```

```python
# transpose the dataframe of tpm
tpm_df_transposed = tpm_reset_index_tpm_without_name_df.T
# create new column calls 'gtex_id' used for mergeing later
tpm_df_transposed['gtex_id'] = tpm_df_transposed.index
tpm_df_transposed
```

```python
#read the GTEx annotations Sample Attribute file as tsv 
annotations_sample_attribute_path = "/private6/Projects/dadush/SIFT_try2/erez_request/GTEx_annotations_SampleAttribute_selected_cols.tsv"
annotations_sample_attribute_df = pd.read_csv(annotations_sample_attribute_path, sep='\t')
annotations_sample_attribute_df
```

```python
# merge by left joining - the origin database is TPM and the sample attribute is joining
merge_df = pd.merge(tpm_df_transposed, annotations_sample_attribute_df, how='left', left_on='gtex_id', right_on='SAMPID')
merge_set_index_drop_df=merge_df.drop('SAMPID', axis=1)
merge_set_index_drop_df
```

```python
#select only brain and liver tissue as differnt data frame
barin_df = merge_set_index_drop_df[(merge_set_index_drop_df['SMTS']=='Brain')]
liver_df = merge_set_index_drop_df[(merge_set_index_drop_df['SMTS']=='Liver')]
liver_df
```

```python
#All gene samples not from liver or brain
all_other_tissue_df = merge_set_index_drop_df[(merge_set_index_drop_df['SMTS']!='Liver')&(merge_set_index_drop_df['SMTS']!='Brain')]
all_other_tissue_df
```

```python
all_other_tissue_df_drop_col=all_other_tissue_df.drop(['SMTS','SMTSD'], axis=1)
all_other_tissue_df_drop_col_reset_ix= all_other_tissue_df_drop_col.set_index(['gtex_id'])
all_other_tissue_df_drop_col_reset_ix
```

```python
all_other_tissue_df_transpose=all_other_tissue_df_drop_col_reset_ix.T
all_other_tissue_df_transpose
```

```python
#Average of all non-brain or liver samples of a particular gene
all_other_tissue_df_transpose['average_tpm_other_tissues'] = all_other_tissue_df_transpose.mean(numeric_only=True, axis=1)
all_other_tissue_df_transpose
```

```python
# Creating a data frame with the name of the gene, the average of the expression and the tissue it came from
tiny_other_tissue_average_tpm_df = pd.DataFrame()
tiny_other_tissue_average_tpm_df['gene_id'] = all_other_tissue_df_transpose.index
tiny_other_tissue_average_tpm_df['average_tpm_other_tissue'] = all_other_tissue_df_transpose['average_tpm_other_tissues'].to_list()
tiny_other_tissue_average_tpm_df['tissue'] = 'Other_tissues'
tiny_other_tissue_average_tpm_df
```

```python
# Throwing out all the rows that have duplicates in the gene name - since we fixed the gene name at the beginning together with the ID, there are no duplicates
tiny_other_tissue_average_tpm_df_drop_dup = tiny_other_tissue_average_tpm_df.drop_duplicates(subset='gene_id')
tiny_other_tissue_average_tpm_df_drop_dup
```

Focusing on creating a table that contains only the genes that are expressed in the brain, when it is enough that in one area of the brain it is expressed above 10 TPM on average

```python
barin_df
```

```python
# Creating a list of all the unique tissues that appear in the brain's database
list_part_of_brain = barin_df.SMTSD.unique().tolist()
```

```python
#Splitting the df according to sub-tissues in the brain
grouped_brain = barin_df.groupby(['SMTSD'])
grouped_brain.get_group('Brain - Amygdala')
```

```python
#loop which keeps in DF all the genes that in at least one tissue had an average higher than 10 TPM 
total_tissue_in_barin_df = df = pd.DataFrame(columns = ['gene_id', 'average_tpm', 'tissue'])
for brain_part in list_part_of_brain:
    part_brain_df = grouped_brain.get_group(brain_part)
    part_brain_df=part_brain_df.drop(['SMTS','SMTSD'], axis=1)
    part_brain_df_reset_ix= part_brain_df.set_index(['gtex_id'])
    part_brain_transpose=part_brain_df_reset_ix.T
    part_brain_transpose['average'] = part_brain_transpose.mean(numeric_only=True, axis=1)
    #Creating a reduced table of the gene name and its average byte level
    tiny_part_brain_over_10_tpm_df = pd.DataFrame()
    tiny_part_brain_over_10_tpm_df['gene_id'] = part_brain_transpose.index
    tiny_part_brain_over_10_tpm_df['average_tpm'] = part_brain_transpose['average'].to_list()
    tiny_part_brain_over_10_tpm_df['tissue'] = 'Brain'
    tiny_part_brain_over_10_tpm_df['sub_tissue'] = brain_part
    # left join between the table of gene expression in the sub-tissues in the brain with the table of expression levels of genes in other tissues
    merged_df = tiny_part_brain_over_10_tpm_df.merge( tiny_other_tissue_average_tpm_df_drop_dup, left_on='gene_id',right_on='gene_id', how='left',suffixes=('_brain', '_other_tissue'))
    total_tissue_in_barin_df=total_tissue_in_barin_df.append(merged_df, ignore_index = True)

    
```

```python
total_tissue_in_barin_df
```

```python
# The selection of the genes whose expression levels in a subtissue in the brain were higher than 10 TPM and in the other tissues were lower than 10 TPM
over_10_tpm_in_brain_and_less_10_tpm_in_other_df = total_tissue_in_barin_df[(total_tissue_in_barin_df['average_tpm']>10)&(total_tissue_in_barin_df['average_tpm_other_tissue']<10)]
over_10_tpm_in_brain_and_less_10_tpm_in_other_df
```

```python
# Rearranging the table
over_10_tpm_in_brain_and_less_10_tpm_in_other_df['gene_name'] = over_10_tpm_in_brain_and_less_10_tpm_in_other_df['gene_id'].str.split('#').str[0] 
over_10_tpm_in_brain_and_less_10_tpm_in_other_df['gene_id'] = over_10_tpm_in_brain_and_less_10_tpm_in_other_df['gene_id'].str.split('#').str[1] 
over_10_tpm_in_brain_and_less_10_tpm_in_other_df
```

```python
over_10_tpm_in_brain_and_less_10_tpm_in_other_df['tissue_brain']= over_10_tpm_in_brain_and_less_10_tpm_in_other_df['sub_tissue']
over_10_tpm_in_brain_and_less_10_tpm_in_other_df=over_10_tpm_in_brain_and_less_10_tpm_in_other_df.drop(['tissue','sub_tissue'], axis=1)
over_10_tpm_in_brain_and_less_10_tpm_in_other_df = over_10_tpm_in_brain_and_less_10_tpm_in_other_df.rename(columns={'average_tpm':'average_tpm_brain'})
over_10_tpm_in_brain_and_less_10_tpm_in_other_df
```

```python
# Design as Rona requested
# Copy selected columns
final_brain_rona_design_df = over_10_tpm_in_brain_and_less_10_tpm_in_other_df[['gene_id','gene_name','average_tpm_brain','tissue_brain','average_tpm_other_tissue']].copy()
# Connecting two columns together and saving 5 digits after the decimal point
final_brain_rona_design_df['tissue_brain'] = final_brain_rona_design_df['tissue_brain'].str.split('- ').str[1]
final_brain_rona_design_df['average_tpm_brain'] = final_brain_rona_design_df['tissue_brain'] + ' - ' + final_brain_rona_design_df['average_tpm_brain'].apply(str).str[:7]
final_brain_rona_design_df = final_brain_rona_design_df.drop('tissue_brain', axis=1)
# final_brain_rona_design_df.rename(columns={'gene_id':'Gene ID','gene_name':'Gene name','average_tpm_liver':'Liver average TPM','average_tpm_other_tissue':'Other average TPM'})
final_brain_rona_design_df
```

```python
# Grouping the table by gene name and ID number and turning the brain tissue columns and their scores into a new column
final_brain_rona_design_df = final_brain_rona_design_df.groupby(['gene_id','gene_name','average_tpm_other_tissue'])['average_tpm_brain'].apply(list).reset_index(name='Brain average TPM')
final_brain_rona_design_df = final_brain_rona_design_df[['gene_id','gene_name','Brain average TPM','average_tpm_other_tissue']]
final_brain_rona_design_df  = final_brain_rona_design_df.rename(columns={'gene_id':'Gene ID','gene_name':'Gene name','average_tpm_other_tissue':'Other average TPM'})
final_brain_rona_design_df
```
