```python
def get_change_in_amino_acid( base_change, three_nuc, loc_in_three_nuc):
        # check if the mutation is a type of base change and not del/ins or something else
        if ">" in base_change:
            # if there is no details on the location of change or this is not an SNV return nan as the alternate three nucleotides
            if (pd.isna(loc_in_three_nuc)  or pd.isna(base_change) or pd.isna(three_nuc) ) :
                return float("nan")
            # if all details are exists
            else:
                # # if the original three nuc that we get from DNA seq are nan, get the three nuc from new_codon_seq column  - information that came from the RNA sequence 
                # if pd.isna(three_nuc):
                #     three_nuc = new_codon_seq
                # # split the name col and take the alternate base
                alt_base = base_change.split(">")[1]
                # convert to list
                three_nuc = list(three_nuc)
                # change the specific postion in three nuc  - that I get the mutation
                three_nuc[int(loc_in_three_nuc)] = alt_base
                # convert to str
                three_nuc = "".join(three_nuc)
                return three_nuc
        else:
            return float("nan")
#apply the funcation get_change_in_amino_acid so that I get the mutant three nuc and assign it in new_alt_three_nuc column
ff_df['new_alt_three_nuc'] = ff_df.apply(lambda x: get_change_in_amino_acid(x.mutation, x.new_codon_seq,x.loc_in_nuc), axis=1)
ff_df
```

```python
# funcation that flat nested list
def flatten(x):
    result = []
    for el in x:
        if hasattr(el, "__iter__") and not isinstance(el, str):
            result.extend(flatten(el))
        else:
            result.append(el)
    return result


def all_edit_for_three_nuc(alt_three_nuc,ref_base,alt_base):
    lst_of_index = [[0, 1, 2],[0, 2], [0, 1], [1, 2],[0],[1],[2]]
    edit_option = []
    for item in lst_of_index:
        word = alt_three_nuc
        # try to edit every position in item
        for position in item:
            if word[position]==ref_base:
                word = word[:position] + alt_base + word[position+1:]
        # If the edited nucleotide tride is not yet in the list and is not equal to the mutant nucleotide tride then add it to the list
        if word not in edit_option and word!=alt_three_nuc:        
            edit_option.append(word)
    return edit_option
# funcation for editing both editor
def edit_on_all_option(alt_three_nuc):
    edit_list = all_edit_for_three_nuc(alt_three_nuc,'A','G')
    lst_both_edit =[]
    for seq_edit_by_apobec in edit_list:
        edit_nuc = all_edit_for_three_nuc(seq_edit_by_apobec,'C','T')
        if edit_nuc not in lst_both_edit and edit_nuc !=seq_edit_by_apobec:        
               lst_both_edit.append(edit_nuc)
    return list(flatten(lst_both_edit))
    
# selects the appropriate function according to the editing enzyme 
if edit_enzyme =='adar':
    ff_df['List_of_editing']= ff_df.apply(lambda x: all_edit_for_three_nuc(x['new_alt_three_nuc'],'A','G'), axis=1)
elif edit_enzyme == 'apobec':
    ff_df['List_of_editing']= ff_df.apply(lambda x: all_edit_for_three_nuc(x['new_alt_three_nuc'],'C','T'), axis=1)
elif edit_enzyme == 'both':
    ff_df['List_of_editing']= ff_df.apply(lambda x: edit_on_all_option(x['new_alt_three_nuc']), axis=1)
ff_df
```

```python
def find_refseq_id(data_farme):
    # Read the file that connect between rna refseq if and their protein refseq id 
    refseq_rna_protein_file = "/private5/Projects/dadush/clinvar_base_editing/tidy_code_check/reseq_rna_id_refseq_protien_id"
    refseq_rna_protein_df = pd.read_csv(refseq_rna_protein_file, sep='\t')
    # for each refseq id find the protein id and assign it in Refseq_protein_id column
    for index, row in data_farme.iterrows():
        name = row['gene_id']
        value =  refseq_rna_protein_df[refseq_rna_protein_df['#hg38.ncbiRefSeqCurated.name']==name]['hg38.ncbiRefSeqLink.protAcc'].values[0]
        data_farme.loc[index,'Refseq_protein_id'] = value
    return data_farme
without_nonesene_adar_ff_df = find_refseq_id(without_nonesene_adar_ff_df)
without_nonesene_adar_ff_df
```

```python

# function that get protein name and the suffix of file and create file
def write_to_file(prot_name,text,suffix_of_file):
    if suffix_of_file == '.fa':
        file_dir ="/private6/Projects/dadush/SIFT_try2/sift6.2.1/tmp/refseq_id_with_fasta_format/" + str(prot_name) +".fa"
    elif suffix_of_file == '_subt':
        file_dir ="/private6/Projects/dadush/SIFT_try2/sift6.2.1/tmp/subtitions_sift_format/" + str(prot_name) +"_subt"
    ofile = open(file_dir, "w")
    ofile.write(text)
    ofile.close()
#function that take only the relelvant part of protein seq and change the location of mutation according to this 
def change_muataion_location_and_seq(change_location,seq_of_fasta):
    # if the change is in the first 250 
    if (change_location-250)< 0:
        new_seq = seq_of_fasta[0:500]
        return change_location,new_seq
    # if the change is in the last 250
    elif (change_location+250)>len(seq_of_fasta):
        new_seq = seq_of_fasta[-500:]
        dis_from_end = len(seq_of_fasta) -  change_location
        new_location= 500-dis_from_end
        return new_location,new_seq
    # if the change is in the middle then cut 250 aa before and after 
    else:
        new_seq = seq_of_fasta[change_location-251:change_location+249]
        #new_location = 250
        new_location = 251
        return new_location,new_seq

def check_length(seq_of_fasta,prot_name,sift_format):
    if len(seq_of_fasta)<500:
        #create a file of subtition
        write_to_file(prot_name,sift_format,'_subt')
        # create a FASTA file for protein seq
        fasta_format=  ">" + str(prot_name) +"\n" +str(seq_of_fasta)+"\n"
        write_to_file(prot_name,fasta_format,'.fa')
    else:
        # loop that create fasta and subtition files for every change 
            change = sift_format
            location = int(change[1:-1])
            new_loc,new_seq = change_muataion_location_and_seq(location,seq_of_fasta)
            new_sift_change_format = change[0]+str(new_loc)+change[-1]
            #new_prot_id = protein_id+"_"+change
            fasta_format =  ">"+prot_name+"\n" +str(new_seq)+"\n"
            write_to_file(prot_name,fasta_format,'.fa')
            write_to_file(prot_name,new_sift_change_format,'_subt')

for index, row in without_nonesene_adar_ff_df.iterrows():
    fasta_id=row['Refseq_protein_id']
    fasta_seq = seq_dict[fasta_id]  
    new_prot_name = str(fasta_id)+"^"+str(row['sift_format_ref_mut'])
    check_length(fasta_seq,new_prot_name,row['sift_format_ref_mut'])
    new_prot_name = str(fasta_id)+"^"+str(row['sift_format_ref_edit'])
    check_length(fasta_seq,new_prot_name,row['sift_format_ref_edit'])
    
    
```

```python
import os
from os import listdir
from os.path import isfile, join
from unittest import result
import pandas as pd
import numpy as np
import json


def read_sift_result(path_of_data,result_path):
    ok_file = 0
    error_file = 0
    # read only files name  without suffix as a list
    suffix = '.fa'
    list_of_files_with_fa = [f[0:-len(suffix)] for f in listdir(path_of_data) if isfile(join(path_of_data, f))]
    #onlyfiles
    # read the results:
    #result_path = "/private6/Projects/dadush/SIFT_try2/sift6.2.1/adar_edit_against_human_4008_xglobal_error/"
    # "/private6/Projects/dadush/SIFT_try2/sift6.2.1/tmp_complete/" 
    list_for_df = []
    for file_name in list_of_files_with_fa:
        #create the new path
        file_name_dir  = result_path + file_name + '.SIFTprediction'
        # check if there is result of prediction 
        if os.path.isfile(file_name_dir):
            mutation_name_list =[]
            #read the result file
            with open(file_name_dir) as f:
                for line in f:
                    line = line.strip()
                    listed_line = [x for x in line.split('\t')]
                    # if line is warrning
                    if len(listed_line)==1: 
                        continue
                    # if the score is "NOT SCORED"
                    elif len(listed_line)==2:
                        new_name = file_name +'-'+listed_line[0]
                        score = listed_line[1]
                        list_for_df.append((new_name,score))
                    # standard result:
                    elif len(listed_line)==6:
                        new_name = file_name +'-'+listed_line[0]
                        score = listed_line[2]
                        list_for_df.append((new_name,score))
                ok_file+=1
                # if ok_file==10:
                print(file_name)
                
        else:
            if '*' in file_name:
                new_name = file_name+'-stopcodon'
                list_for_df.append((new_name,0))  
            #print(file_name)
            else:
                new_name = file_name+'-no prediction'
                list_for_df.append((new_name,"no prediction"))
            error_file+=1
    # create dataframe from list
    df_of_results = pd.DataFrame(list_for_df,columns=['mutation','score_human'])
    print ("Number of results without prediction:")
    print(len(df_of_results[df_of_results['score_human']=="no prediction"]))
    print ("Number of results NOT SCORED:")
    print(len(df_of_results[df_of_results['score_human']=='NOT SCORED']))

    # print(ok_file)
    # print(error_file)
    return df_of_results
```

```python
# organize the data of results
def organize_result_df(df_of_results):
    #separate the exchange to other column called 'split_location'
    df_of_results['split_location'] =  df_of_results['mutation'].str.split('^').str[1]
    df_of_results['location_of_mutation'] = df_of_results['split_location'].str.split('-').str[0]
    # df_of_results.loc[df_of_results["split_location"].str.contains('-')& ~ df_of_results["split_location"].str.contains('_') , "location_of_mutation"] = df_of_results['split_location'].str.split('-').str[1]
    # df_of_results.loc[df_of_results["split_location"].str.contains('-')& df_of_results["split_location"].str.contains('_') , "location_of_mutation"] = df_of_results['split_location'].str.split('-').str[0].str.split('_').str[1]
    #extract refseq id to other column called 'refseq_id'
    df_of_results['refseq_id'] =  df_of_results['mutation'].str.extract('([N]{1}[P]{1}[_]{1}\d+[\.]+\d{1})', expand=True)
    return df_of_results



    
```

```python
def add_sift_results_to_original_df(original_df,organized_reult_df):
    for index, row in original_df.iterrows():
        try:
            mut_score = organized_reult_df.loc[(organized_reult_df['refseq_id']==row['Refseq_protein_id'])& (organized_reult_df['location_of_mutation']==row['sift_format_ref_mut']),'score_human']
            edit_score = organized_reult_df.loc[(organized_reult_df['refseq_id']==row['Refseq_protein_id'])& (organized_reult_df['location_of_mutation']==row['sift_format_ref_edit']),'score_human']
        
            if mut_score.empty:
                original_df.loc[index,'sift_score_ref_mut']= np.nan
            else:
                original_df.loc[index,'sift_score_ref_mut']= mut_score.iloc[0]
            if edit_score.empty:
                original_df.loc[index,'sift_score_ref_edit']= np.nan
            else:
                original_df.loc[index,'sift_score_ref_edit'] = edit_score.iloc[0]
        except Exception as e:
            print(e)
```

```python
# add column of number of editing events
def count_editing(seq1,seq2):
   count = sum(1 for a, b in zip(seq1, seq2) if a != b)
   return count
final_with_max_score_df['Number of editing'] = final_with_max_score_df.apply(lambda x: count_editing(x['List_of_editing'],x['new_alt_three_nuc']), axis=1)
final_with_max_score_df
```
