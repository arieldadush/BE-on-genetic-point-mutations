```python
# import libraries 
import pandas as pd
import numpy as np
import subprocess
import re
```

```python
#A list of the mutations that we can fix
list_of_mutation = ['G>A','C>T','A>G','T>C']
#A function that prints the amount of mutations there are of each type that we can edit
def count_amount_of_mutations(df,mutations_list):
    for mutation in mutations_list:
        mutation_df = df[df['name']==mutation]
        print(mutation+':')
        print(len(mutation_df))
```

Creating the FF database from the source, from the original clinvar database that was downloaded on 08/11/21

```python
#read the clinvar databse file as tsv  
clinvar_path = "/private5/Projects/dadush/clinvar_base_editing/clinvarMain.bed"
col_names = ['chrom','chromStart','chromEnd','name','score','strand','thickStart','thickEnd','reserved','blockCount','blockSizes','chromStarts','origName','clinSign','reviewStatus','type','geneId','molConseq','snpId','nsvId','rcvAcc','testedInGtr','phenotypeList','phenotype','origin','assembly','cytogenetic','_jsonHgvsTable','_hgvsProt','numSubmit','lastEval','guidelines','otherIds','_mouseOver','_clinSignCode','_originCode','_allTypeCode','_varLen','_starCount','_variantId','_dbVarSsvId']
clinvar_df = pd.read_csv(clinvar_path, sep='\t',names=col_names)
pd.set_option('display.max_columns', None)
# count the mutations that can edit
count_amount_of_mutations(clinvar_df,list_of_mutation)
clinvar_df
```

```python
# show the distribution of the types of mutations in the database
clinvar_df.groupby(['type'])['type'].count()
```

```python
# first filter - The selection of mutations that are SNV 
snv_df = clinvar_df[clinvar_df['type']=='single nucleotide variant']
count_amount_of_mutations(snv_df,list_of_mutation)
snv_df
```

```python
# show the distribution of the molecular consequence
snv_df.groupby(['molConseq'])['molConseq'].count()
```

```python
snv_df[snv_df['molConseq'].isna()]
```

```python
# second filter - Selecting only the mutations that are within genes
part_df = snv_df[(snv_df['molConseq']!='genic downstream transcript variant') & (snv_df['molConseq']!='genic upstream transcript variant')&(snv_df['molConseq']!='no sequence alteration')&~snv_df['molConseq'].isna()]
count_amount_of_mutations(part_df,list_of_mutation)
part_df
```

```python
# check if there is nan value in molConseq column
part_df[part_df['molConseq'].isna()]
```

```python
# show the distribution of the point mutations that exist in order to find problematic ones
part_df.groupby(['name'])['name'].count()
```

```python
# third filter - Selection of only mutations in which there is a true replacement of one nucleotide with another
part1_df = part_df[part_df['name'].str.contains('>')]
part2_df = part1_df[~part1_df['name'].str.contains('c.')]
part3_df = part2_df[part2_df['name']!='G>G']
count_amount_of_mutations(part3_df,list_of_mutation)
part3_df
```

```python
# show the distribution of the point mutations that exist in order to prove that it is ok now
part3_df.groupby(['name'])['name'].count()
```

```python
# check if there is nan value in name column
part3_df[part3_df['name'].isna()]
```

```python
# Fourth filter - elimination of mutations found in the mitochondrial chromosome
part3_df = part3_df[part3_df['chrom']!='chrM' ]
count_amount_of_mutations(part3_df,list_of_mutation)
part3_df
```

```python
# check distribution of chrom
part3_df.groupby(['chrom'])['chrom'].count()
```

```python
# check if there is nan value in chrom column
part3_df[part3_df['chrom'].isna()]
```

```python
# Fifth filter - selecting the mutations that are defined as pathogenic or those that are in conflict according to the database we downloaded from the UCSC but have at least one submitter that defined them as pathogenic in the VCF
clinvar_snv_pathogenic_df = part3_df[ part3_df['clinSign'].str.contains('pathogenic',flags=re.IGNORECASE, regex=True)]
count_amount_of_mutations(clinvar_snv_pathogenic_df,list_of_mutation)
clinvar_snv_pathogenic_df
```

Reading the VCF table downloaded from the clinvar website on 08/11/21. The purpose of the table is to provide additional information that is missing in the UCSC database

new merge

```python
# extract the clinical significant from INFO col 
merge_2_without_nan_df['CLNSIG'] = merge_2_without_nan_df['INFO'].str.split('CLNSIG=').str[1].str.split(';').str[0]
# Creating a column for all the rowss that are Conflicting interpretations of pathogenicity, so that we can identify the distribution of the submitters for each of them
merge_2_without_nan_df.loc[merge_2_without_nan_df["CLNSIG"].str.contains('Conflicting_interpretations_of_pathogenicity', na=False), "CLNSIGCONF" ] = merge_2_without_nan_df['INFO'].str.split('CLNSIGCONF=').str[1].str.split(';').str[0]

# Creating a column that if there is at least one submitter that claims it is pathogenic or potentially pathogenic it will be considered pathogenic
merge_2_without_nan_df.loc[(merge_2_without_nan_df["CLNSIGCONF"].str.contains('Pathogenic\(', na=False))|(merge_2_without_nan_df["CLNSIGCONF"].str.contains('Likely_pathogenic\(', na=False)), "pathogenic_at_least_one_subbmiter" ] = 'yes'
# If there is not at least one submitter who claims that it is pathogenic or potentially pathogenic, then it will be considered not
merge_2_without_nan_df.loc[~(merge_2_without_nan_df["CLNSIGCONF"].isna())&~(merge_2_without_nan_df["CLNSIGCONF"].str.contains('Pathogenic\(', na=False))&~(merge_2_without_nan_df["CLNSIGCONF"].str.contains('Likely_pathogenic\(', na=False)), "pathogenic_at_least_one_subbmiter" ] = 'no'

# Selecting all lines that are defined as pathogenic according to the UCSC and in addition the lines that are defined in conflict according to the UCSC but have at least one submitter that defined them as pathogenic in the VCF
# Erez request
new_snv_final_path_df = merge_2_without_nan_df[(merge_2_without_nan_df["clinSign"].str.contains('Conflicting')&((merge_2_without_nan_df['pathogenic_at_least_one_subbmiter']=='yes')))|~(merge_2_without_nan_df["clinSign"].str.contains('Conflicting'))]
count_amount_of_mutations(new_snv_final_path_df,list_of_mutation)
new_snv_final_path_df
```

```python
# Actual garbage value filling, to maintain correct placement of values when we run bash scripts 
new_snv_final_path_df = new_snv_final_path_df.fillna('nAn')
new_snv_final_path_df
```
