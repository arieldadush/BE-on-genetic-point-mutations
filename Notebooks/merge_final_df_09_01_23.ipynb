select the best editor to each mutation:
1. maximum SIFT score for every mutation
2. the edit with the minimum of editing events
Finally, if there is a mutation that still appears twice, it is unlikely that it can be edited by several enzymes, so we leave both of them to allow the user to choose which way he prefers.
*In our case, there remained two such lines, each of which had two options of editing enzymes.

```python
# check if there is a mutation that appears twice and we would like to choose for it the enzyme that edits best
pd.Series(total_df['origName']).is_unique
total_df[total_df['origName'].duplicated()]
```

```python
total_df[total_df.index.duplicated()]
```

```python
#select the maximum score for every mutation
idx = total_df.groupby(['origName'])['sift_score_ref_edit'].transform(max) == total_df['sift_score_ref_edit']
final_with_max_score_df = total_df[idx]
final_with_max_score_df
```

```python
# check if there is a mutation that appears twice and we would like to choose for it the enzyme that edits best
final_with_max_score_df[final_with_max_score_df['origName'].duplicated()]
```

```python
# add column of number of editing events
def count_editing(seq1,seq2):
   count = sum(1 for a, b in zip(seq1, seq2) if a != b)
   return count   
final_with_max_score_df['Number of editing'] = final_with_max_score_df.apply(lambda x: count_editing(x['List_of_editing'],x['new_alt_three_nuc']), axis=1)
#final_with_max_score_df
# take only the edit with the minimum of editing events
idx = final_with_max_score_df.groupby(['origName'])['Number of editing'].transform(min) == final_with_max_score_df['Number of editing']
final_with_max_score_min_edit_df = final_with_max_score_df[idx]
final_with_max_score_min_edit_df
```

```python
# check if there is a mutation that appears twice and we would like to choose for it the enzyme that edits best
final_with_max_score_min_edit_df[final_with_max_score_min_edit_df['origName'].duplicated()]
```
