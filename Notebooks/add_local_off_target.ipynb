```python
# read thr RNA seqs file
from Bio import SeqIO
input_file = "/private5/Projects/dadush/clinvar_base_editing/cds_fasta"
#input_file = "/private5/Projects/dadush/clinvar_base_editing/cds_utrs.fasta"
#fasta_sequences = SeqIO.parse(open(input_file),'fasta')
seq_dict = {rec.id.replace('hg38_ncbiRefSeqCurated_', '')+rec.description : rec.seq for rec in SeqIO.parse(open(input_file), "fasta")} #if ('v1_' not in rec.description) or ('_alt' not in rec.description) or ('_fix' not in rec.description)}
seq_dict
```

```python
[key for key, value in seq_dict.items() if 'NM_001372044.2' in key]
```

```python
# remove all alternative seqs
#compare_dict = seq_dict.copy()
for key in seq_dict.copy():
    if 'v1_' in key :
        seq_dict.pop(key)
    elif '_alt' in key:
        seq_dict.pop(key)
    elif '_fix' in key:
        seq_dict.pop(key)
    elif 'v2_' in key:
        seq_dict.pop(key)
    else:
        new_key = key.split('hg38')[0]
        seq_dict[new_key] = seq_dict.pop(key)
len(seq_dict)       
```

```python
# def that add exon seq to df
import re
def get_exon_seq(loc_in_exon,refseq_id):
    if pd.isna(loc_in_exon):
        return np.nan
    elif not loc_in_exon.isdigit():
        return np.nan
    else:
        #location =  re.sub("[^0-9]", "",loc_in_exon)
        location  = int(loc_in_exon)
        #for fasta in fasta_sequences:
        #fasta.id = fasta.id.replace('hg38_ncbiRefSeqCurated_', '')
        if refseq_id in seq_dict.keys():
            seq = seq_dict[refseq_id]
            if (location+10 > len(seq)) and (location-11<0):
                seq = seq[0:len(seq)]
            elif location+10 > len(seq):
                seq = seq[location-11:len(seq)]
            elif location-11<0:
                seq = seq[0:location+10]
            else:
                seq = seq[location-11:location+10]
            return str(seq)
        else:
            # refseq_id in str(fasta.id):
            #     seq=fasta.seq[int(location)-21:int(location)+20] 
            #     return str(seq)
            return np.nan       
# get 10 RNA bases before and after the mutation
final_table_df['RNA_10_bases_before_and_after'] = final_table_df.apply(lambda x: get_exon_seq(x.location_on_exon, x.gene_id),axis=1)
final_table_df
```

```python
from Bio.Seq import Seq
from Bio.SeqUtils import seq3
# A function that accepts the codon and returns the name of the amino acid in a three letter format
def get_amino_acid(codon):
    # Checking whether this is an instance of a string
    if isinstance(codon, str):
        tran_cod = Seq(codon).translate()
        #seq_three =  seq3(tran_cod)
        return str(tran_cod)
    else:
        return np.nan
# translate the seq before edit
final_table_df['AA_seq_before_edit'] =final_table_df.apply(lambda x: get_amino_acid(x.RNA_10_bases_before_and_after_frame), axis=1)
final_table_df
```

```python
# translate the seq after edit
final_table_df['AA_seq_after_edit'] =final_table_df.apply(lambda x: get_amino_acid(x.RNA_local_off_target_edit_frame_edit), axis=1)
final_table_df
```

```python

# A function that accepts the codon and returns the name of the amino acid in a three letter format
def get_dif_in_seq(s1,s2):
    # Checking whether this is an instance of a string
    if isinstance(s1, str) and isinstance(s2, str):
        count=sum(c1!=c2 for c1,c2 in zip(s1,s2))
        return count
    else:
        return np.nan
# count the mismatch in amino acid sew between before and after the edit local off target
final_table_df['total_mismatch_in_aa_seq'] =final_table_df.apply(lambda x: get_dif_in_seq(x.AA_seq_before_edit,x.AA_seq_after_edit), axis=1)
final_table_df
```
